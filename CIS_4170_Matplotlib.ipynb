#Scatterplot


```python
#Adding title and labels to scatterplot
df.plot(x="Height", y ="Weight", title ='scatterplot showing height vs weight', kind = "scatter")

plt.show()
```

#Pie Chart


```python
#Putting the weights on a pie chart
plot = df.plot.pie(y='Weight', figsize=(5, 5))
```

#Bar Chart


```python
#Adding Title, labels, color and symbols to the bar chart
df.plot(x="Height", y = "Weight", title="Bar Plot Showing Height vs Weight", kind="bar",color='g',hatch='o')

plt.show()
```

#Histogram

```python
#Adding title, label, color, legend to the histogram
bin_number = 75

plt.hist(Weight,
         bin_number, 
         range=(100, 300), 
         histtype='stepfilled', 
         align='mid', 
         color='orange', 
         label='Weight')
plt.legend()
plt.title('Weight Histogram')
plt.show()
```

```python
bin_number = 75

plt.hist(Height,
         bin_number, 
         range=(5.0, 7.0), 
         histtype='stepfilled', 
         align='mid', 
         color='orange', 
         label='Height')
plt.legend()
plt.title('Height Histogram')
plt.show()
```

#Line Chart

```python
#Adding title, label, legend, annotation to the line chart
df.plot(x="Height", y = "Weight", title="Line Plot Showing Height vs Weight", kind="line",color='g')
plt.xlabel="Height"
plt.ylabel="Weight"
plt.legend(['Height', 'Weight'], loc=0)
plt.annotate(xy=[6, 250], s='Heavy and Tall')
plt.annotate(xy=[5.6, 140], s='Skinny and Short')
ax = plt.gca()
ax.grid(True)
plt.show()
```

### **Subplots**

```python
#Adding label, title to the subplots.
import matplotlib.pylab as plt
fig, axes=plt.subplots(nrows=1,ncols=2, figsize=(15,5))
df['Height'].plot(ax=axes[0])
df['Weight'].plot(ax=axes[1])
axes[0].set_title('Height of the Players')
axes[0].set_label('Index')
axes[1].set_title('Weight of the Players')
axes[1].set_label('Index')

;
```
